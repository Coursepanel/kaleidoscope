```python
# Define a function to remove the patterns
def remove_patterns(x):
    if isinstance(x, str):
        x = re.sub(r"\d+\. ?", "", x)  # Remove patterns like "1."
        x = re.sub(r"\d+\)", "", x)  # Remove patterns like "1)"
        x = re.sub(r"\\t", "", x)  # Remove tab characters
        x = x.strip()
    return x



df = df.applymap(remove_patterns)
```

```python
pattern = r"\b[A-Za-z]{2}\d{4}\b"

extraction_func = lambda x:", ".join(re.findall(pattern, x)) 
df['extractedCoursePrereqCodes'] = df['prerequisitesOfCurrentCourse'].apply(extraction_func)
df.head(5)
```

```python
def convert_string_to_array(value):
    array = []
    if(type(value) == str):
        array = [s.strip() for s in value.split("; ")]
    return array
```

```python
def convert_string_to_array_2(value):
    array = []
    if(type(value) == str):
        array = [s.strip() for s in value.split(", ")]
    return array
```

```python
def return_cypher_for_row(row, index):
    discipline = row['deptCode']
    id = row['id']
    courseName = row['name']
    courseContent = row['courseContent'] 
    description = row['description'] 
    courseType = row['courseType']
    courseCode = row['courseCode']
    tagsList = convert_string_to_array(row["tags"])
    textBooksList = convert_string_to_array_2(row["textBooks"])
    referenceBooksList = convert_string_to_array_2(row["referenceBooks"])
    extractedPrereqCourseCodesList = [] 
    if(row["extractedPrereqCourseCodes"]):
        extractedPrereqCourseCodesList = convert_string_to_array_2(row["extractedPrereqCourseCodes"])
    
    base_value = index
    course = f'c{base_value}'
    p = f'p{base_value}'
    t = f't{base_value}'
    b = f'b{base_value}'
    bx = f'bx{base_value}'
    dept = f'd{base_value}'
    r1 = f'r1{base_value}'
    r2 = f'r2{base_value}'
    r3 = f'r3{base_value}'
    r4 = f'r4{base_value}'
    r5 = f'r5{base_value}'
    c1 = f'''
        MERGE ({dept}:Department{{name:'{discipline}'}}) 
        MERGE ({course}:Course
            {{
                id:'{id}',
                name:'{courseName}', 
                extractedPrereqCourseCodes: {extractedPrereqCourseCodesList}
            }}
        )'''
    c2 = f"MERGE ({dept})<-[{r1}:BELONGS_TO]-({course})"
    c3 = f'''
        WITH {course} 
        UNWIND {tagsList} AS tag
        WITH {course}, tag
        MERGE ({t}:Tag{{name:tag}})
    '''
    c4 = f"MERGE ({t})<-[{r2}:TAG]-({course})"
    c5 = f'''
        WITH {course} 
        UNWIND {textBooksList} AS book
        WITH {course}, book
        MERGE ({b}:Book{{name:book}})
    '''
    c6 = f"MERGE ({b})<-[{r3}:BOOK]-({course})"
    c7 = f'''
        WITH {course} 
        UNWIND {referenceBooksList} AS book
        WITH {course}, book
        MERGE ({bx}:Book{{name:book}})
    '''
    c8 = f"MERGE ({bx})<-[{r4}:BOOK]-({course})"

    return (c1+c2+c3+c4+c5+c6+c7+c8).replace("\n", "").strip()
```

```python
### Generating Cypher code from pandas dataframe
row = df.iloc[0]
return_cypher_for_row(row, 0)
```

```python
# Open a file for writing
with open('code19.txt', 'w') as f:
    # Iterate over the rows of the DataFrame
    for i, (index, row) in enumerate(df.iterrows()):
        # if(index<10):
            # Write the values to the file, separated by a tab character
        f.write(str(return_cypher_for_row(row, i)) + '\n')
```

```python
def return_prereq_cypher(row, index):
    extractedPrereqCourseCodesList = convert_string_to_array_2(row["extractedPrereqCourseCodes"])
    id = row['id']
    
    base_value = index
    course = f'c{base_value}'
    r6 = f'r6{base_value}'
    epc = f'epc{base_value}'

    c0 = f'''MATCH ({course}:Course{{id:{id}}})'''
    c1 = f'''
        WITH {course} 
        UNWIND {extractedPrereqCourseCodesList} AS prereq
        WITH {course}, prereq
        MATCH ({epc}:Course{{courseCode:prereq}})
    '''
    c2 = f"MERGE ({epc})<-[{r6}:PREREQUISITE]-({course})"

    return (c0+c1+c2).replace("\n", "").strip()
```

```python
### Generating Cypher code from pandas dataframe
row = df.iloc[0]
return_prereq_cypher(row, 0)
```

```python
#### After all the course nodes with their respective courseCodes have been created in the Neo4j graph, we can link prerequisites to each other
with open('prereq_code18.txt', 'w') as f:
    # Iterate over the rows of the DataFrame
    for i, (index, row) in enumerate(df.iterrows()):
        # if(index<10):
            # Write the values to the file, separated by a tab character
        f.write(str(return_prereq_cypher(row, i)) + '\n')
```
